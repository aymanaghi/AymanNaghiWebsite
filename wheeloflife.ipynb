```python
import matplotlib.pyplot as plt
import numpy as np

def plot_wheel_of_life(categories, scores, title="Wheel of Life",
                       line_color='gray', fill_color='skyblue', bg_color='white'):
    """
    Plots a Wheel of Life visualization with customizable colors.

    Args:
    - categories: A list of life area categories.
    - scores: A list of corresponding scores for each category (between 0 and 10).
    - title: Optional title for the plot.
    - line_color: Color for the lines connecting the center to the scores.
    - fill_color: Color for filling the area enclosed by the lines and scores.
    - bg_color: Background color for the plot.
    """

    N = len(categories)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False)
    angles = np.append(angles, angles[0])
    scores = np.append(scores, scores[0])

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    fig.patch.set_facecolor(bg_color) # Set background color

    ax.plot(angles, scores, color=line_color, linewidth=1, linestyle='solid')
    ax.fill(angles, scores, color=fill_color, alpha=0.4)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=10)
    ax.set_ylim(0, 10)
    ax.grid(True)
    ax.set_title(title, fontsize=14)

    plt.show()

# Updated categories and example scores
categories = ['Work', 'Family', 'Friends', 'Romance', 'Sex', 'Creativity', 'Learning', 'Finances', 'Spirit','Gaming']
scores = [10, 3, 3, 8, 4, 9, 9, 9, 9, 7]  # Replace with your own scores

# Example usage with custom colors
plot_wheel_of_life(categories, scores, line_color='blue', fill_color='lightgreen', bg_color='lightyellow')
```
